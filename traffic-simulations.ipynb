```python
from traffic_simulation import Road, Simulation
import numpy as np
import statistics as st
import random
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
time_car_trials = []
num_cars = 30
sim_time = 60
road_length = 1000
car_position_arr = np.linspace(0, road_length-1, num=num_cars, dtype=int)

sim1000results = []
for i in range(10000):
    road = Road(road_length)
    sim = Simulation(num_cars, car_position_arr, sim_time, road)
    sim1000results.append(sim.run())
#print(time_car_trials)

cars_per_sec_array = sim.get_np_array()
```

```python
st.mean(sim1000results[0])
```

```python
st.pstdev(sim1000results[0])
```

```python
np.shape(cars_per_sec_array)
```

```python
for i in range(0,60):
    plt.scatter([i]*len(cars_per_sec_array[i]), cars_per_sec_array[i, :])
plt.show()
```
